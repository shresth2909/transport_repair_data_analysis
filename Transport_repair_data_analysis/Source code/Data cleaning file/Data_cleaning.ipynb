```python
# Cell 2: Loading the CSV file 

file_path = 'F:\Mytraining\DataAnalystBootcamp\Python\Data analysis with python\Axion Ray- Assesement\Transport_repair.csv'
# Cell 4: Advanced handling for difficult files
try:
    # First read as binary then decode
    with open(file_path, 'rb') as f:
        content = f.read().decode('latin1')  # or other encoding that works
    
    # Now read as CSV from string
    from io import StringIO
    df = pd.read_csv(StringIO(content))
    print("Successfully read file with manual decoding")
except Exception as e:
    print(f"Failed to read file: {e}")

df = pd.read_csv(file_path, encoding='latin1')
```

```python
# Display basic information about the dataframe
print("Dataframe Info:")
print(df.info())

# Display first 5 rows
print("\nFirst 5 rows:")
display(df.head())

# Display summary statistics for numerical columns
print("\nSummary Statistics:")
display(df.describe(include='all'))
```

```python
# Cell 3: Handle missing values
# Display count of missing values per column
print("Missing values per column before handling:")
print(df.isnull().sum())

# Strategy 1: Drop columns with too many missing values (adjust threshold as needed)
threshold = len(df) * 0.7  # 70% threshold
df = df.dropna(thresh=threshold, axis=1)

# Strategy 2: Impute missing values
# For numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
if len(num_cols) > 0:
    num_imputer = SimpleImputer(strategy='median')
    df[num_cols] = num_imputer.fit_transform(df[num_cols])

# For categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
if len(cat_cols) > 0:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# Verify missing values after handling
print("\nMissing values per column after handling:")
print(df.isnull().sum())
```

```python
# Cell 4: Clean categorical columns
if len(cat_cols) > 0:
    # Standardize capitalization (convert all to lowercase)
    for col in cat_cols:
        df[col] = df[col].astype(str).str.lower().str.strip()
    
    # Handle specific known inconsistencies (add more as needed)
    # Example: Fixing typos in a 'gender' column
    if 'gender' in df.columns:
        df['gender'] = df['gender'].replace({'m': 'male', 'f': 'female', 'malee': 'male'})
    
    # Display unique values after cleaning
    print("\nUnique values in categorical columns after cleaning:")
    for col in cat_cols:
        print(f"{col}: {df[col].unique()}")
else:
    print("No categorical columns found.")
```

```python
# Cell 5: Clean numerical columns and handle outliers
if len(num_cols) > 0:
    # Convert columns to proper numeric type (handling errors)
    for col in num_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Handle outliers using IQR method (optional)
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Cap outliers
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    
    print("\nNumerical columns cleaned and outliers handled.")
else:
    print("No numerical columns found.")
```

```python
# Cell 7: Save the cleaned data to a new CSV file
cleaned_file_path = 'cleaned_data.csv'
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")

# Display final dataframe info
print("\nFinal dataframe info:")
print(df.info())
```
